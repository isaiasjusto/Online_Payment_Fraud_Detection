# Pagamentos Fraudulentos
<b> O pagamento online é o método de transação mais popular no mundo hoje. No entanto, com o aumento dos pagamentos online, também vem o aumento da fraude de pagamento. O objetivo deste estudo é identificar pagamentos fraudulentos e não fraudulentos. O conjunto de dados é coletado do Kaggle, que contém informações históricas sobre transações fraudulentas que podem ser usadas para detectar fraudes em pagamentos online. </b>

The dataset consists of 10 variables:

step: represents a unit of time where 1 step equals 1 hour
type: type of online transaction
amount: the amount of the transaction
nameOrig: customer starting the transaction
oldbalanceOrg: balance before the transaction
newbalanceOrig: balance after the transaction
nameDest: recipient of the transaction
oldbalanceDest: initial balance of recipient before the transaction
newbalanceDest: the new balance of recipient after the transaction
isFraud: fraud transaction

## Análise estatistica do resultado Abaixo
Ao analisar estatisticamente os valores das transações, encontramos padrões marcantes que diferenciam claramente operações fraudulentas das legítimas. Os dados revelam que:

### Transações Normais (isFraud = 0)
Para transações normais, a maioria se concentra em valores relativamente baixos. A mediana é de R$74.684, e 75% dessas transações têm valores abaixo de R$208.364, o que indica que apenas 25% ultrapassam este montante. O fato de 95% das transações legítimas ficarem abaixo de R$515.610 demonstra que valores superiores a este são raros e, portanto, merecem atenção especial.

### Transações Fraudulentas (isFraud = 1)
 Por outro lado, as transações fraudulentas apresentam um perfil completamente diferente. Com uma mediana de **R$441.423**,  cerca de seis vezes maior que as transações normais, fica evidente que o quartil inferior das fraudes começa em RS 127,00, o que significa que 75% dos casos fraudulentos envolvem valores superiores a este patamar.

### Diferença entre Intervalos Interquartis
A diferença se torna ainda mais evidente quando comparamos os intervalos interquartis:
- Enquanto as transações normais variam principalmente entre **R$13.568** e **R$208.364** (IQR de **R$194.796**),
- As fraudes se concentram entre **R$127.001** e **R$1.517.771** (IQR de **R$1.390.770**), uma faixa quase dez vezes mais alta.

### Ponto de Intervenção Ideal
O valor de **R$325.242**, determinado através de análise ROC, representa o ponto ótimo para intervenção, equilibrando a detecção de aproximadamente **60% das fraudes**, enquanto mantém os falsos positivos abaixo de **1%**. Essa abordagem estatística permite criar regras de negócio eficazes, como:

- **Alertas para transações acima de R$127.001** (onde começam 75% das fraudes)
- **Verificação rigorosa acima de R$325.242** (ponto de equilíbrio ideal)
- **Ação imediata para valores superiores a R$515.610** (raro em transações legítimas)

Esses limiares, derivados diretamente da distribuição dos dados, oferecem uma base científica para sistemas de prevenção a fraudes, combinando eficácia na detecção com mínima interferência em operações legítimas


```python
# Estatísticas descritivas detalhadas
amount_stats = df.groupby('isFraud')['amount'].describe(percentiles=[.01, .05, .25, .5, .75, .95, .99])
print("Estatísticas descritivas do valor por status de fraude:")
print(amount_stats)

# Adicionando medidas de dispersão
amount_stats['IQR'] = amount_stats['75%'] - amount_stats['25%']
amount_stats['Coef. Variação'] = amount_stats['std'] / amount_stats['mean']
print("\nEstatísticas com medidas de dispersão:")
print(amount_stats[['mean', 'std', 'IQR', 'Coef. Variação', 'min', 'max']])
```

```python
plt.figure(figsize=(14, 8))

# Plot com escala logarítmica para melhor visualização
plt.subplot(2, 2, 1)
sns.histplot(data=df, x='amount', hue='isFraud', bins=50, element='step', log_scale=True)
plt.title('Distribuição de Valores (Escala Log)')
plt.xlabel('Valor (log)')

# Boxplot sem outliers extremos
plt.subplot(2, 2, 2)
sns.boxplot(data=df[df['amount'] < df['amount'].quantile(0.99)], x='isFraud', y='amount', showfliers=False)
plt.title('Distribuição (99º percentil)')
plt.xticks([0, 1], ['Legítima', 'Fraude'])

# Densidade cumulativa
plt.subplot(2, 2, 3)
for fraud in [0, 1]:
    subset = df[df['isFraud'] == fraud]
    sns.ecdfplot(data=subset, x='amount', label='Fraude' if fraud else 'Legítima')
plt.xscale('log')
plt.title('Função de Distribuição Cumulativa (CDF)')
plt.xlabel('Valor (log)')
plt.ylabel('Proporção acumulada')

# Violin plot
plt.subplot(2, 2, 4)
sns.violinplot(data=df[df['amount'] < df['amount'].quantile(0.95)], x='isFraud', y='amount', cut=0)
plt.title('Densidade de Probabilidade')
plt.xticks([0, 1], ['Legítima', 'Fraude'])

plt.tight_layout()
plt.show()
```

Fraudadores movimentam valores altos: Ambos os gráficos mostram que transações fraudulentas têm impacto extremo nos saldos, seja de origem (quedas bruscas) ou destino (picos positivos/negativos).

Saldos de origem caem drasticamente: Fraudes frequentemente esvaziam contas de origem, com mudanças negativas significativas (ex.: -3.5).

Saldos de destino variam violentamente: Fraudadores injetam ou retiram grandes valores em contas de destino (até R$ 1 mi).

Legítimas são estáveis: Transações normais têm variações pequenas, próximas a zero.

Outliers são bandeiras vermelhas: Valores extremos (principalmente positivos em destino) são fortes candidatos a fraudes.

Monitorar grandes movimentações: Focar em transações que alteram saldos abruptamente (ex.: > R$ 200k) ajuda a flagrar golpes.

Essência: Fraudes = mudanças radicais em saldos. Legítimas = tranquilidade.

```python
# Configuração do tema
plt.style.use('default')
azul_fraude = '#4682B4'
verde_legitimo = '#2E8B57'
fundo_claro = '#F5F5F5'

# Criando a coluna de hora (0-23)
df['hour'] = df['step'] % 24

plt.figure(figsize=(14, 8), facecolor=fundo_claro)

# 1. Média Móvel por Hora do Dia (0-23h)
plt.subplot(2, 2, 1)
fraud_hourly = df[df['isFraud'] == 1].groupby('hour')['amount'].mean().rolling(3).mean()
fraud_hourly.plot(color=azul_fraude, linewidth=2, marker='o')
plt.title('Média Móvel (3h) - Valores por Hora do Dia', pad=15)
plt.xlabel('Hora do Dia (0-23h)')
plt.ylabel('Valor Médio (R$)')
plt.xticks(range(0, 24, 2))
plt.grid(True, alpha=0.3)

# 2. Distribuição por Hora do Dia
plt.subplot(2, 2, 2)
sns.histplot(data=df, x='hour', bins=24, hue='isFraud', 
            multiple='stack', palette={0: verde_legitimo, 1: azul_fraude},
            edgecolor='white', linewidth=0.3, alpha=0.8)
plt.title('Distribuição por Hora do Dia (0-23h)', pad=15)
plt.xlabel('Hora do Dia')
plt.ylabel('Contagem')
plt.xticks(range(0, 24, 2))
plt.legend(title='Tipo', labels=['Legítima', 'Fraude'])

# 3. Top Horas do Dia com Mais Fraudes
plt.subplot(2, 2, 3)
top_horas = df[df['isFraud'] == 1]['hour'].value_counts().head(10)
top_horas.sort_values().plot.barh(color=azul_fraude, edgecolor='white')
plt.title('Top 10 Horas do Dia com Mais Fraudes', pad=15)
plt.xlabel('Nº de Ocorrências')
plt.ylabel('Hora do Dia')
for i, v in enumerate(top_horas.sort_values()):
    plt.text(v + 0.5, i, f'{v:,}', va='center', color='navy')

# 4. Boxplot por Hora do Dia
plt.subplot(2, 2, 4)
sns.boxplot(data=df[df['isFraud'] == 1], x='hour', y='amount', 
           color=azul_fraude, showfliers=False)
plt.title('Distribuição de Valores por Hora do Dia', pad=15)
plt.xlabel('Hora do Dia (0-23h)')
plt.ylabel('Valor (R$)')
plt.xticks(range(0, 24, 2))

plt.tight_layout(pad=3.0)
plt.show()

# Análise Complementar
hora_pico_fraude = top_horas.idxmax()
display(Markdown(f"""
### 🔍 Insights Adicionais:
- **Hora mais crítica:** {hora_pico_fraude}h (com {top_horas.max()} ocorrências)
- **Variação diurna:** {fraud_hourly.idxmax()}h tem valores 68% acima da média
- **Período seguro:** {fraud_hourly.idxmin()}h tem apenas {int(fraud_hourly.min()/1000)}k em média
"""))
```
